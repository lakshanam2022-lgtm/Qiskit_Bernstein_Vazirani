```python
# ================================================================
# ğŸ§  Bernsteinâ€“Vazirani Algorithm â€” Setup Cell
# ================================================================

!pip install qiskit qiskit-aer qiskit-ibm-runtime matplotlib --quiet

from qiskit import QuantumCircuit, transpile
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator
from qiskit_aer.noise import NoiseModel, depolarizing_error
from qiskit_ibm_runtime import QiskitRuntimeService
import matplotlib.pyplot as plt

print("âœ… Qiskit environment ready!")
```

```python
# ================================================================
# ğŸ§© Task 1: Change the secret string s and verify the output
# ================================================================

s = '1011'     # ğŸ”¹ Change this to test other secret strings
n = len(s)

# Create the Bernsteinâ€“Vazirani circuit
bv_circuit = QuantumCircuit(n + 1, n)
bv_circuit.x(n)              # Ancilla |1>
bv_circuit.h(range(n + 1))   # Apply Hadamard to all qubits

# Oracle for f(x) = sÂ·x (mod 2)
for i, bit in enumerate(s):
    if bit == '1':
        bv_circuit.cx(i, n)

bv_circuit.h(range(n))
bv_circuit.measure(range(n), range(n))

# Simulate
simulator = AerSimulator()
compiled = transpile(bv_circuit, simulator)
result = simulator.run(compiled).result()
counts = result.get_counts()

print("âœ… Secret string s =", s)
print("Measured output:", counts)

# --- Text-based visualization (Colab Safe) ---
print("\nQuantum Circuit:")
print(bv_circuit.draw("text"))
plot_histogram(counts)
plt.show()
```

```python
# ================================================================
# ğŸ§© Task 2: Modify the oracle to include constant bit b
# ================================================================

b = 1   # Try 0 or 1
bv_circuit_b = QuantumCircuit(n + 1, n)
bv_circuit_b.x(n)
bv_circuit_b.h(range(n + 1))

# Oracle: f(x) = sÂ·x âŠ• b
for i, bit in enumerate(s):
    if bit == '1':
        bv_circuit_b.cx(i, n)
if b == 1:
    bv_circuit_b.x(n)   # Apply constant bit b

bv_circuit_b.h(range(n))
bv_circuit_b.measure(range(n), range(n))

result_b = simulator.run(transpile(bv_circuit_b, simulator)).result()
counts_b = result_b.get_counts()

print(f"Modified oracle with constant bit b = {b}")
print("Measured output:", counts_b)

print("\nQuantum Circuit:")
print(bv_circuit_b.draw("text"))
plot_histogram(counts_b)
plt.show()
```

```python
# ================================================================
# ğŸ§© Task 4: Add noise using qiskit_aer.noise.NoiseModel
# ================================================================

p_error = 0.05  # Adjust noise probability
noise_model = NoiseModel()
noise_model.add_all_qubit_quantum_error(depolarizing_error(p_error, 1), ['h', 'x'])
noise_model.add_all_qubit_quantum_error(depolarizing_error(p_error, 2), ['cx'])

noisy_sim = AerSimulator(noise_model=noise_model)
result_noisy = noisy_sim.run(transpile(bv_circuit, noisy_sim)).result()
counts_noisy = result_noisy.get_counts()

print(f"Noisy simulation with depolarizing error p = {p_error}")
plot_histogram([counts, counts_noisy], legend=['Ideal', 'Noisy'])
plt.show()
```
